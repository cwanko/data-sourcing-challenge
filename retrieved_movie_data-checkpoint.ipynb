```python
# Create an empty list to store the reviews
reviews_list = []

# Loop through pages 0-19
for page in range(20):
    # Create query with a page number
    nyt_base_url = "https://api.nytimes.com/svc/search/v2/articlesearch.json"
    nyt_params = {
        'fq': 'section_name:("Movies") AND type_of_material:("Review") AND headline:("love")',
        'sort': 'newest',
        'fl': 'headline,web_url,snippet,source,keywords,pub_date,byline,word_count',
        'begin_date': '20130101',
        'end_date': '20230531',
        'page': page,
        'api-key': nyt_api_key
    }

    # Make a "GET" request and retrieve the JSON
    response = requests.get(nyt_base_url, params=nyt_params)
    
    if response.status_code == 200:
        reviews = response.json()
        
        # Add a twelve second interval between queries to stay within API query limits
        time.sleep(12)
        
        try:
            # Save the reviews to the reviews_list
            for doc in reviews["response"]["docs"]:
                reviews_list.append(doc)
            
            # Print the page that was just retrieved
            print(f"Checked page {page}")
            
            # If no results are found on the page, break the loop
            if not reviews["response"]["docs"]:
                print(f"No results on page {page}")
                break
        except KeyError:
            print(f"No results on page {page}")
            break
    else:
        print(f"Error: {response.status_code}")
        print(response.text)  # Print the error response for more details
        break



```

```python
# Extract 'name' and 'value' from items in "keywords" column
def extract_keywords(keyword_list):
    extracted_keywords = ""
    for item in keyword_list:
        # Extract 'name' and 'value'
        keyword = f"{item['name']}: {item['value']};" 
        # Append the keyword item to the extracted_keywords list
        extracted_keywords += keyword
    return extracted_keywords

# Fix the "keywords" column by converting cells from a list to a string
reviews_df['keywords'] = reviews_df['keywords'].apply(extract_keywords)
reviews_df
```

### Access The Movie Database API

```python
# Prepare The Movie Database query
url = "https://api.themoviedb.org/3/search/movie?query="
tmdb_key_string = "&api_key=" + tmdb_api_key

# Complete URL example
#query = "Inception"  # Example movie query
#complete_url = url + requests.utils.quote(query) + tmdb_key_string

# Print only the query portion of the URL for debugging purposes
#print("Query URL: " + url + requests.utils.quote(query))

```

```python
# Create an empty list to store the results
tmdb_movies_list = []

# Create a request counter to sleep the requests after a multiple of 50 requests
request_counter = 0

# Function to search for a movie on TMDB
def search_movie_on_tmdb(title):
    query_url = url + requests.utils.quote(title) + tmdb_key_string
    response = requests.get(query_url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code} for title: {title}")
        print(response.text)  # Print the error response for more details
        return None

# Loop through the titles
for title in titles_list:
    # Check if we need to sleep before making a request
    if request_counter > 0 and request_counter % 50 == 0:
        print("Sleeping for 60 seconds to respect API rate limits...")
        time.sleep(60)

    # Add 1 to the request counter
    request_counter += 1

    # Perform a "GET" request for The Movie Database
    tmdb_result = search_movie_on_tmdb(title)

    # Include a try clause to search for the full movie details.
    # Use the except clause to print out a statement if a movie is not found.
    try:
        # Get movie id
        movie_id = tmdb_result['results'][0]['id']

        # Make a request for the full movie details
        details_url = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key={tmdb_api_key}"
        details_response = requests.get(details_url)

        # Execute "GET" request with url
        if details_response.status_code != 200:
            print(f"Error: {details_response.status_code} for movie ID: {movie_id}")
            continue
        movie_details = details_response.json()

        # Extract the genre names into a list
        genres = [genre['name'] for genre in movie_details['genres']]

        # Extract the spoken_languages' English name into a list
        languages = [language['english_name'] for language in movie_details['spoken_languages']]

        # Extract the production_countries' name into a list
        production_countries = [country['name'] for country in movie_details['production_countries']]

        # Add the relevant data to a dictionary and append it to the tmdb_movies_list list
        movie_data = {
            'title': title,
            'tmdb_id': movie_id,
            'genres': genres,
            'languages': languages,
            'production_countries': production_countries,
            'release_date': movie_details['release_date'],
            'overview': movie_details['overview']
        }
        tmdb_movies_list.append(movie_data)

        # Print out the title that was found
        print(f"Found {title}")

    except (IndexError, KeyError) as e:
        print(f"{title} not found.")
    
    if request_counter % 200 == 0:
        print("Sleeping at 200 requests")
        time.sleep(60)

# Optional: Save the TMDB movie details to a JSON file for further analysis
with open('tmdb_movie_details.json', 'w') as file:
    json.dump(tmdb_movies_list, file, indent=4)

# Convert the tmdb_movies_list to a DataFrame for further processing if needed
tmdb_df = pd.DataFrame(tmdb_movies_list)

# Preview the first 5 results in DataFrame format
print("Preview of the first 5 TMDB movie details in DataFrame format:")
print(tmdb_df.head())

# Save the DataFrame to a CSV file
tmdb_df.to_csv('tmdb_movie_details.csv', index=False)
```

### Merge and Clean the Data for Export

```python

# Merge the DataFrames on the title column
merged_df = pd.merge(tmdb_df, reviews_df, on='title')

# Display the merged DataFrame
display(merged_df)
```

```python
# List of columns that need fixing
columns_to_fix = ['genres', 'languages', 'production_countries']

# List of characters to remove
chars_to_remove = ["[", "]", "'", '"']

# Loop through the list of columns to fix
for column in columns_to_fix:
    # Convert the column to type 'str'
    merged_df[column] = merged_df[column].astype(str)
    
    # Loop through characters to remove
    for char in chars_to_remove:
        merged_df[column] = merged_df[column].str.replace(char, '',regex=False)

# Display the fixed DataFrame
print("Fixed DataFrame:")
display(merged_df.head())


```

```python

# Drop the "byline.person" column
merged_df = merged_df.drop(columns=['byline.person'])

```

```python
# Delete duplicate rows and reset index
merged_df = merged_df.drop_duplicates().reset_index(drop=True)

# Display the fixed DataFrame
print("Fixed DataFrame:")
display(merged_df.head())

```
